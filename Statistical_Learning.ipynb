```r
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# SECTION 0: INSTALL AND LOAD REQUIRED LIBRARIES
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# This section ensures all necessary packages are installed and loaded.
# If a package is not found, it will be installed from CRAN.

if (!require("ggplot2")) install.packages("ggplot2")
if (!require("corrplot")) install.packages("corrplot")
if (!require("dplyr")) install.packages("dplyr")
if (!require("caret")) install.packages("caret")
if (!require("cluster")) install.packages("cluster")
if (!require("caTools")) install.packages("caTools")
if (!require("randomForest")) install.packages("randomForest")
if (!require("xgboost")) install.packages("xgboost")
if (!require("pROC")) install.packages("pROC")
if (!require("reshape2")) install.packages("reshape2")

library(ggplot2)
library(corrplot)
library(dplyr)
library(caret)
library(cluster)
library(caTools)
library(randomForest)
library(xgboost)
library(pROC)
library(reshape2)

# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# SECTION 1: DATA LOADING AND EXPLORATORY DATA ANALYSIS (EDA)
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# In this step, we load the dataset, inspect its structure, handle missing values,
# and visualize the data distributions and correlations.

# 1.1: Load Data
df <- read.csv("/content/credit_risk_dataset.csv")

# 1.2: Initial Inspection
cat("--- First 6 rows of the dataset ---\n")
print(head(df))

cat("\n--- Structure of the dataset ---\n")
str(df)

cat("\n--- Summary statistics ---\n")
summary(df)

# 1.3: Handle Missing Values
cat("\n--- Missing values before imputation ---\n")
print(sapply(df, function(x) sum(is.na(x))))

# Impute missing values with the column median
df$person_emp_length[is.na(df$person_emp_length)] <- median(df$person_emp_length, na.rm = TRUE)
df$loan_int_rate[is.na(df$loan_int_rate)] <- median(df$loan_int_rate, na.rm = TRUE)

cat("\n--- Missing values after imputation ---\n")
print(sapply(df, function(x) sum(is.na(x))))

# 1.4: Data Visualization
# Histograms for numerical variables
numerical_cols <- sapply(df, is.numeric)
par(mfrow=c(3, 3)) # Arrange plots in a 3x3 grid
for(col_name in names(df[, numerical_cols])) {
  hist(df[[col_name]], main=paste("Distribution of", col_name), xlab=col_name, col='skyblue')
}
par(mfrow=c(1, 1)) # Reset plot layout

# Bar plots for categorical variables
categorical_cols <- c("person_home_ownership", "loan_intent", "loan_grade", "cb_person_default_on_file")
for(col in categorical_cols) {
  print(
    ggplot(df, aes_string(x = col)) +
    geom_bar(fill = "steelblue") +
    ggtitle(paste("Distribution of", col)) +
    theme_minimal()
  )
}

# Correlation Matrix Heatmap
correlation_matrix <- cor(df[, numerical_cols])
corrplot(correlation_matrix, method = "color", type = "upper", order = "hclust",
         addCoef.col = "black", tl.col = "black", tl.srt = 45, title = "Correlation Matrix", mar=c(0,0,1,0))

# Boxplots for outlier detection
par(mfrow=c(3, 3))
for(col_name in names(df[, numerical_cols])) {
  boxplot(df[[col_name]], main=paste("Boxplot of", col_name), col='lightblue')
}
par(mfrow=c(1, 1))


# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# SECTION 2: FEATURE ENGINEERING AND CREATION
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Here, we convert categorical variables to a numeric format, create a new
# feature, and scale all features for modeling.

# 2.1: One-Hot Encoding for Categorical Variables
target_variable <- df$loan_status
df_features <- df[, !(names(df) %in% c("loan_status"))]

dummies <- dummyVars(" ~ .", data = df_features)
df_processed <- data.frame(predict(dummies, newdata = df_features))

df_processed$loan_status <- target_variable # Add target back

# 2.2: Feature Creation: income_loan_ratio
# Handle cases where loan_amnt might be zero to avoid division by zero
df_processed$loan_amnt[df_processed$loan_amnt == 0] <- 1e-9
df_processed$income_loan_ratio <- df_processed$person_income / df_processed$loan_amnt

# 2.3: Feature Scaling (Standardization)
# We exclude the binary target variable 'loan_status' from scaling.
cols_to_scale <- names(df_processed)[names(df_processed) != "loan_status"]
preproc_values <- preProcess(df_processed[, cols_to_scale], method = c("center", "scale"))
df_processed[cols_to_scale] <- predict(preproc_values, df_processed[, cols_to_scale])

cat("\n--- First 6 rows of the fully processed dataset ---\n")
print(head(df_processed))


# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# SECTION 3: PRINCIPAL COMPONENT ANALYSIS (PCA)
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# We apply PCA to reduce the dimensionality of the dataset and visualize the
# explained variance to choose the right number of components.

# 3.1: Apply PCA
# Separate features before applying PCA
X_features <- df_processed[, !(names(df_processed) %in% c("loan_status"))]
pca_result <- prcomp(X_features, center = FALSE, scale. = FALSE) # Data is already scaled

# 3.2: Analyze Explained Variance
explained_variance_ratio <- summary(pca_result)$importance[2, ]
cumulative_explained_variance <- summary(pca_result)$importance[3, ]

# Plot the cumulative explained variance
plot(cumulative_explained_variance, xlab = "Number of Components", ylab = "Cumulative Explained Variance",
     type = "b", main = "Cumulative Explained Variance vs. Number of Components", col = "blue")
abline(h = 0.95, col = "red", lty = 2) # 95% variance threshold

# 3.3: Transform data with the selected number of components (e.g., 15)
n_components <- 15
df_pca <- as.data.frame(pca_result$x[, 1:n_components])

cat("\n--- First 6 rows of the PCA-transformed dataset ---\n")
print(head(df_pca))


# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# SECTION 4: K-MEANS CLUSTERING
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Using the PCA-reduced data, we perform K-Means clustering to identify
# potential segments in the data. The optimal number of clusters is found
# using the Elbow Method.

# 4.1: Find Optimal K using the Elbow Method
wss <- (nrow(df_pca)-1) * sum(apply(df_pca, 2, var))
for (i in 2:10) wss[i] <- sum(kmeans(df_pca, centers=i, nstart=25)$withinss)

plot(1:10, wss, type="b", xlab="Number of Clusters (K)",
     ylab="Within groups sum of squares", main="Elbow Method for Optimal K", col="darkgreen")

# 4.2: Apply K-Means
# Based on the plot, k=3 seems to be a reasonable elbow point.
optimal_k <- 3
set.seed(42)
kmeans_result <- kmeans(df_pca, centers = optimal_k, nstart = 25)

# 4.3: Analyze Clusters
# Add cluster labels back to the original, unscaled dataframe for interpretability
df$cluster <- kmeans_result$cluster

cluster_analysis <- df %>%
  group_by(cluster) %>%
  summarise(across(where(is.numeric), list(mean = mean, median = median)),
            person_home_ownership_mode = names(which.max(table(person_home_ownership))),
            loan_intent_mode = names(which.max(table(loan_intent))),
            loan_grade_mode = names(which.max(table(loan_grade))),
            .groups = 'drop')

cat("\n--- Analysis of Clusters ---\n")
print(as.data.frame(cluster_analysis))


# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# SECTION 5: MODELING
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# We split the data and train three different classification models:
# Logistic Regression, Random Forest, and XGBoost.

# 5.1: Prepare Data and Split into Training/Testing sets
X <- df_processed[, !(names(df_processed) %in% c("loan_status"))]
y <- as.factor(df_processed$loan_status)

set.seed(42)
split <- sample.split(y, SplitRatio = 0.8)
X_train <- subset(X, split == TRUE)
X_test <- subset(X, split == FALSE)
y_train <- subset(y, split == TRUE)
y_test <- subset(y, split == FALSE)

cat("\n--- Data Split ---\n")
cat("X_train shape:", dim(X_train), "\n")
cat("X_test shape:", dim(X_test), "\n")

# 5.2: Train Logistic Regression
train_data_log <- cbind(X_train, loan_status = y_train)
log_reg <- glm(loan_status ~ ., data = train_data_log, family = binomial)

# 5.3: Train Random Forest
rand_forest <- randomForest(x = X_train, y = y_train, ntree = 100, random.state = 42)

# 5.4: Train XGBoost
# Convert data to DMatrix format required by XGBoost
y_train_numeric <- as.numeric(as.character(y_train))
y_test_numeric <- as.numeric(as.character(y_test))
dtrain <- xgb.DMatrix(data = as.matrix(X_train), label = y_train_numeric)
dtest <- xgb.DMatrix(data = as.matrix(X_test), label = y_test_numeric)

params <- list(objective = "binary:logistic", eval_metric = "auc", eta = 0.1, seed = 42)
xgb_model <- xgb.train(params, dtrain, nrounds = 100, watchlist = list(eval = dtest, train = dtrain), print_every_n = 50, early_stopping_rounds = 10)

cat("\n--- All models trained successfully! ---\n")

# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# SECTION 6: MODEL PERFORMANCE EVALUATION
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# We evaluate the performance of each model on the test set using various
# classification metrics.

# 6.1: Predictions
pred_prob_log <- predict(log_reg, X_test, type = "response")
pred_class_log <- as.factor(ifelse(pred_prob_log > 0.5, 1, 0))

pred_class_rf <- predict(rand_forest, X_test)
pred_prob_rf <- predict(rand_forest, X_test, type = "prob")[,2]

pred_prob_xgb <- predict(xgb_model, dtest)
pred_class_xgb <- as.factor(ifelse(pred_prob_xgb > 0.5, 1, 0))

# 6.2: Confusion Matrices and Metrics
levels(y_test) <- c("0", "1") # Ensure levels match for confusion matrix
cm_log <- confusionMatrix(pred_class_log, y_test)
cm_rf <- confusionMatrix(pred_class_rf, y_test)
cm_xgb <- confusionMatrix(pred_class_xgb, y_test)

# 6.3: ROC-AUC Scores
roc_log <- roc(y_test, pred_prob_log)
roc_rf <- roc(y_test, pred_prob_rf)
roc_xgb <- roc(y_test, pred_prob_xgb)

# 6.4: Compile Performance Metrics into a DataFrame
performance_metrics <- data.frame(
  Model = c("Logistic Regression", "Random Forest", "XGBoost"),
  Accuracy = c(cm_log$overall['Accuracy'], cm_rf$overall['Accuracy'], cm_xgb$overall['Accuracy']),
  Precision = c(cm_log$byClass['Pos Pred Value'], cm_rf$byClass['Pos Pred Value'], cm_xgb$byClass['Pos Pred Value']),
  Recall = c(cm_log$byClass['Sensitivity'], cm_rf$byClass['Sensitivity'], cm_xgb$byClass['Sensitivity']),
  F1_Score = c(cm_log$byClass['F1'], cm_rf$byClass['F1'], cm_xgb$byClass['F1']),
  ROC_AUC = c(auc(roc_log), auc(roc_rf), auc(roc_xgb))
)

cat("\n--- Model Performance Comparison ---\n")
print(performance_metrics)

# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# SECTION 7: FEATURE IMPORTANCE
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# We extract and visualize which features were most influential for the
# Random Forest and XGBoost models.

# 7.1: Random Forest Feature Importance
importance_rf <- as.data.frame(importance(rand_forest))
importance_rf$Feature <- rownames(importance_rf)
importance_rf <- importance_rf[order(-importance_rf$MeanDecreaseGini), ]

print(
  ggplot(head(importance_rf, 15), aes(x = reorder(Feature, MeanDecreaseGini), y = MeanDecreaseGini)) +
  geom_bar(stat = "identity", fill = "darkturquoise") +
  coord_flip() +
  ggtitle("Top 15 Features - Random Forest") +
  xlab("Features") +
  ylab("Importance (Mean Decrease in Gini)")
)

# 7.2: XGBoost Feature Importance
importance_xgb <- xgb.importance(model = xgb_model)
print(xgb.plot.importance(head(importance_xgb, 15)))

# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# SECTION 8: FINAL RESULTS AND COMPARISON
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Finally, we present a summary and visual comparison of the model performances.

# 8.1: Melt data for plotting with ggplot2
performance_melted <- melt(performance_metrics, id.vars = "Model")

# 8.2: Create comparison plot
print(
  ggplot(performance_melted, aes(x = Model, y = value, fill = Model)) +
  geom_bar(stat = "identity", position = "dodge") +
  facet_wrap(~variable, scales = "free_y") +
  geom_text(aes(label = round(value, 3)), vjust = 1.5, color = "white", size = 3.5) +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  ggtitle("Model Performance Comparison Across Metrics") +
  ylab("Score") +
  xlab("Model")
)

# 8.3: Summary
cat("\n--- ANALYSIS SUMMARY ---\n")
best_model <- performance_metrics[which.max(performance_metrics$ROC_AUC), ]
cat(paste("Based on ROC-AUC, the best performing model is", best_model$Model, "with a score of", round(best_model$ROC_AUC, 4), ".\n"))
cat("Both Random Forest and XGBoost significantly outperformed Logistic Regression across all metrics.\n")
cat("Key predictive features for both tree-based models include income-to-loan ratio, loan percentage of income, and home ownership status (specifically renting).\n")

# --- END OF SCRIPT ---
```
